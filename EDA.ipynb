```python
column_list=list(train_data.columns)
object_list=[]
integer_list=[]


for i in column_list:
    if train_data[i].dtype=="object":
        object_list.append(i)
    elif train_data[i].dtype=="int64":
        integer_list.append(i)
    elif train_data[i].dtype=="float64":
        integer_list.append(i)
    
object_list,integer_list
```

```python
isnull_object_column_list=[]
isnull_integer_column_list=[]
for i in object_list:
    if train_data[i].isnull().sum()>0:
        print(f"The {i} columns total isnull value:      {train_data[i].isnull().sum()}")
        isnull_object_column_list.append(i)
print("The other object columns isnull value is 0")
print("############################################################################")
for i in integer_list:
    if train_data[i].isnull().sum()>0:
        print(f"The {i} columns total isnull value:      {train_data[i].isnull().sum()}")
print("The other integer columns isnull value is 0")
  
```

```python
total = train_data.isnull().sum().sort_values(ascending=False)
percent = (train_data.isnull().sum()/train_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```
